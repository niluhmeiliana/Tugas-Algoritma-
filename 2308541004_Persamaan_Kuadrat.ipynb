```python
print("============================= Project 2 =============================")
# 1.Memanggil numpy
import numpy as np
# 2.Memanggil pyplot
import matplotlib.pyplot as plt
# 3.Memanggil cmath
import cmath
# 4.Masukkan untuk nilai a,b,c
a_04 = float(input("Masukkan nilai dari a "))
b_04 = float(input("Masukkan nilai dari b "))
c_04 = float(input("Masukkan nilai dari c "))
if a_04 != 0:
  print("Persamaan kuadrat yang terbentuk adalah")
else:
  print("Tidak dapat membentuk persamaan kuadrat")
print("Persamaan kuadrat yang dibentuk ({}x**2+{}x+{})".format(a_04,b_04,c_04))
print("Persamaan Kuadratnya adalah ({}x**2+{}x+{})".format(a_04,b_04,c_04))
# 5.Masukkan rumus diskriminan
D_04 = (b_04**2-4*a_04*c_04)
if D_04 == 0:
  print("Akar-akarnya kembar")
# 6.Masukkan rumus x1,x2
  x1_04 = (-b_04+cmath.sqrt(D_04))/(2*a_04)
  x2_04 = (-b_04-cmath.sqrt(D_04))/(2*a_04)
  print("Akar-akar dari persamaannya adalah ({:.1f},{:.1f})".format(x1_04,x2_04))
elif D_04 > 0:
  print("Akar real berbeda")
  x1_04 = (-b_04 + cmath.sqrt(D_04))/(2*a_04)
  x2_04 = (-b_04 - cmath.sqrt(D_04))/(2*a_04)
  print("Akar-akarnya adalah ({:.1f},{:.1f})".format(x1_04,x2_04))
elif D_04 < 0:
  print("Akarnya imajiner")
  x1_04 = (-b_04 + cmath.sqrt(D_04))/(2*a_04)
  x2_04 = (-b_04 - cmath.sqrt(D_04))/(2*a_04)
  print("Akar-akarnya adalah ({:.1f},{:.1f})".format(x1_04,x2_04))
else:
  print("Tidak terdapat solusi pada penyelesaian ini ")

print("================================== Terima Kasih ==================================")
```

```python
print("============================== Project 3 ==============================  ")
# Grafik pada persamaan kuadrat
# Langkah-langkahnya adalah :
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
# 4. Masukkan nilai a, b dan c dalam persamaan ax**2 + bx + c = 0
a_04 = float(input("Nilai konstanta a adalah: "))
b_04 = float(input("Nilai konstanta b adalah: "))
c_04 = float(input("Nilai konstanta c adalah: "))

if a_04 == 0:
    print("Error, nilai yang anda input bukan persamaan kuadrat")
else:
    print("Nilai yang anda input bisa membentuk persamaan kuadrat")
    print("Persamaan Kuadratnya adalah ({}x**2 + {}x + {})".format(a_04, b_04, c_04))

    # 5. Masukkan rumus diskriminan
    D_04 = b_04**2 - 4*a_04*c_04
    if D_04 == 0:
        print("Wahh, Akar-akarnya kembar")
    elif D_04 > 0:
        print("Akar real berbeda")
    else:
        print("Akarnya imajiner")

    # 6. Masukkan rumus x1, x2
    x1_04 = (-b_04 + cmath.sqrt(D_04)) / (2*a_04)
    x2_04 = (-b_04 - cmath.sqrt(D_04)) / (2*a_04)
    print("Akar-akarnya adalah ({:.2f}, {:.2f})".format(x1_04.real, x2_04.real))

    # 7. Format garis
    x_04 = np.linspace(-10, 10, 400)
    y_04 = a_04*x_04**2 + b_04*x_04 + c_04
    plt.figure(figsize=(8, 6))
    plt.plot(x_04, y_04, label=f'{a_04}x^2 + {b_04}x + {c_04} = 0')

    if D_04 >= 0:
        plt.scatter([x1_04.real, x2_04.real], [0, 0], color='red')
        plt.text(x1_04.real, 0.5, f'({x1_04.real:.2f}, 0)', fontsize=12, ha='center')
        plt.text(x2_04.real, 0.5, f'({x2_04.real:.2f}, 0)', fontsize=12, ha='center')
    else:
        plt.scatter([x1_04.real, x2_04.real], [x1_04.imag, x2_04.imag], color='red')
        plt.text(x1_04.real, x1_04.imag, f'({x1_04.real:.2f}, {x1_04.imag:.2f})', fontsize=12, ha='center')
        plt.text(x2_04.real, x2_04.imag, f'({x2_04.real:.2f}, {x2_04.imag:.2f})', fontsize=12, ha='center')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x_04')
    plt.ylabel('y_04')
    plt.legend()
    plt.show()

    print("================================== Terima Kasih ==================================")
```

```python
print("############################## Menentukan Titik Puncak dan Sumbu Simetri ##############################")
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
# 4. Masukkan nilai a,b,c
a_04 = float(input("Nilai a adalah: "))
b_04 = float(input("Nilai b adalah: "))
c_04 = float(input("Nilai c adalah: "))

def hitung_titik_puncak_dan_sumbu_simetri(a_04, b_04, c_04):
    # Sumbu simetri (x) adalah -b / (2a)
    sumbu_simetri = -b_04 / (2 * a_04)

    # Titik puncak (x, y)
    # x dari titik puncak adalah sumbu simetri
    # y dari titik puncak adalah nilai fungsi kuadrat di x = sumbu_simetri
    titik_puncak_x_04 = sumbu_simetri
    titik_puncak_y_04 = a_04 * (titik_puncak_x_04 ** 2) + b_04 * titik_puncak_x_04 + c_04

    return (titik_puncak_x_04, titik_puncak_y_04), sumbu_simetri

def plot_persamaan_kuadrat(a_04, b_04, c_04):
    # Menghitung titik puncak dan sumbu simetri
    titik_puncak, sumbu_simetri = hitung_titik_puncak_dan_sumbu_simetri(a_04, b_04, c_04)

    # Membuat rentang x untuk plot
    x_04 = np.linspace(titik_puncak[0] - 10, titik_puncak[0] + 10, 400)
    y_04 = a_04 * x_04**2 + b_04 * x_04 + c_04

    # Plot grafik persamaan kuadrat
    plt.figure(figsize=(10, 6))
    plt.plot(x_04, y_04, label=f'{a_04}x^2 + {b_04}x + {c_04}')

    # Plot titik puncak
    plt.plot(titik_puncak[0], titik_puncak[1], 'ro')  # Titik puncak berwarna merah
    plt.annotate(f'Titik puncak\n({titik_puncak[0]:.2f}, {titik_puncak[1]:.2f})',
                 xy=titik_puncak,
                 xytext=(titik_puncak[0] + 1, titik_puncak[1] + 10),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    # Plot sumbu simetri
    plt.axvline(x=sumbu_simetri, color='gray', linestyle='--', label='Sumbu simetri')

    # Menambahkan label dan judul
    plt.xlabel('x_04')
    plt.ylabel('y_04')
    plt.title('Titik Puncak Grafik Persamaan Kuadrat')
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.legend()
    plt.grid(True)
    plt.show()

# Menjalankan fungsi plot_persamaan_kuadrat dengan nilai a, b, dan c yang diberikan
plot_persamaan_kuadrat(a_04, b_04, c_04)
print("================================== Terima Kasih ==================================")
```
