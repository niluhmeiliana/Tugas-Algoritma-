```python
import numpy as np
import matplotlib.pyplot as plt
print("################# Program Persamaan Garis #################")
# Persamaan garis pertama melalui titik (w,x) dan (y,z)
# 1.Masukkan nilai titik w,x,y dan z
w_04=1 #Titik(w,x)
x_04=1
y_04=2 #Titik(y,z)
z_04=2
# 2.Menghitung gradien persamaan 1
m1=(z_04-x_04)/(y_04-w_04)
print("Gradien dari garis yang pertama adalah {}".format(m1))
print("Persamaan garis yang pertama adalah y={}x+{} ".format(m1,-m1*w_04+x_04))
#y1 = m1*(x-w_04)+x_04

# Persamaan garis pertama kedua melalui titik (a,b) dan (c,d)
# 3.Masukkan nilai titik a,b,c dan d
a_04=1 #Titik(a,b)
b_04=0
c_04=2 #Titik(c,d)
d_04=1
# 4.Menghitung gradien persamaan 2
m2=(d_04-b_04)/(c_04-a_04)
print("Gradien dari garis yang ke dua adalah {}".format(m2))
print("Persamaan garis yang kedua adalah {}".format(m2,-m2*a_04+b_04))
#y2=m2*(x-a_04)+b_04

if m1==m2:
  print(" Wahhh ternyata gradiennya sama,jadi kemungkinan tidak ada solusi atau solusinya banyak.")
else:
  # Titik potong kedua garis pada (tp_x,tp_y)
# 5.Mencari titik potong
  tp_x = (m1*w_04-x_04-m2*a_04+b_04)/(m1-m2)
  tp_y = m2*(tp_x-w_04)+x_04
  print("Titik potong ({},{}).format(tp_x,tp_y)")
# 6.Menentukan range dan figure
  x=np.arange(tp_x-10,tp_x+10,0.01)
  plt.figure(figsize=(4,4),dpi=200)
# 7.Menampilkan hasil
  plt.plot(x,m1*(x-w_04)+x_04,color= 'red',label='y={}+{}'.format(m1,-m1*w_04+x_04))
  plt.plot(x,m2*(x-a_04)+b_04,color= 'green',label='y={}+{}'.format(m2,-m2*a_04+b_04))
  plt.plot(tp_x,tp_y,color='blue',marker='o',markersize=5)
  plt.legend(loc='best')
  plt.show()
  print("################################# Sekian Terima Kasih #################################")
```
